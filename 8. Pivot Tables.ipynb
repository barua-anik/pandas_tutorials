```python
# pivot simply turns rows - colomn

df.pivot(index="day", columns = "city")
```

# Pivot table: used to summarize and aggregate data 

```python
df = pd.read_csv("weather2.csv")
df
```

```python
df.pivot_table(index="city",columns="date") # will show the mean by default
```

```python
df.pivot_table(index="city",columns="date",aggfunc="sum")
```

```python
df.pivot_table(index="city",columns="date",aggfunc="count")
```

```python
df.pivot_table(index="city",columns="date",aggfunc="diff")
```

```python
df.pivot_table(index="city",columns="date",margins=True)
```

```python
df = pd.read_csv("weather3.csv")
df
```

```python
# converting date from str to date type
df['date'] = pd.to_datetime(df['date'])
df
```

```python
# average temperature in may and in december
df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
```
