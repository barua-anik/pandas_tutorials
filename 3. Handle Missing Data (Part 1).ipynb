```python
import pandas as pd
df = pd.read_csv("weather_data2.csv", parse_dates=["day"])

type(df.day[0]) # date is string type, so we need to convert the type
df
```

```python
df.set_index('day',inplace=True) # setting day as index
df
```

# fillna

```python
# Filling all NaN values
new_df = df.fillna(0)
new_df
```

```python
# Filling some NaN values
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'No event'
})

new_df
```

```python
# fillig the zero values with pervious values 

new_df = df.fillna(method="ffill") # fill forward the values 
new_df
```

```python
# fillig the zero values based on next values 

new_df = df.fillna(method="bfill") # fill back the values 
new_df
```

```python
# fillig the zero values based on next row 

new_df = df.fillna(method="bfill", axis = "columns") # fill back the values 
new_df
```

```python
# fillig the zero values with pervious values with a limit 

new_df = df.fillna(method="ffill", limit=1) # fill forward the values just 1 time
new_df
```

# Interpolate

```python
# Interpolate
# fills the missing value with the avg values, by default linear

new_df = df.interpolate() 
new_df
```

```python
# Interpolate
# fills the missing value with the time values

new_df = df.interpolate(method="time") 
new_df

```

# Dropna

```python
# delete all the rows which have at least 1 Nan

new_df = df.dropna()
new_df
```

```python
# Deleting whole row if all elememnts are NaN

new_df = df.dropna(how="all")
new_df
```

```python
# thresh means how many valid values are needed to keep the data 

new_df = df.dropna(thresh=2)
new_df
```

```python
# Inserting missing date
     
dt = pd.date_range("2017-01-01","2017-01-11") 

idx = pd.DatetimeIndex(dt)
df = new_df.reindex(idx)
df
```
