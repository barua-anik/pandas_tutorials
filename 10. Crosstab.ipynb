```python
import pandas as pd
```

```python
df = pd.read_excel("survey.xls")
df
```

```python
# crosstab shows the frequency... 

pd.crosstab(df.Nationality, df.Sex) # row, column value
```

```python
pd.crosstab(df.Nationality, df.Sex, margins=True) # margins=True shows the total
```

```python
pd.crosstab(df.Nationality, [df.Sex, df.Age], margins=True) 
```

```python
pd.crosstab([df.Sex,df.Nationality], df.Handedness, margins=True) 
```

```python
pd.crosstab([df.Sex,df.Nationality], df.Handedness, normalize="index")  # shows the percentage
```

```python
pd.crosstab([df.Sex], [df.Handedness])
```

```python
import numpy as np
# avg age of male/female being left/right handed
pd.crosstab([df.Sex], [df.Handedness], values =df.Age, aggfunc=np.average)
```
