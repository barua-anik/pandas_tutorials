```python
import pandas as pd
```

```python
df = pd.read_csv("apple_data.csv")
df.head(15)
```

# Working with data with no dates 

```python
df_no_date =  df.drop(columns="Date")
df_no_date.head(15)
```

```python
# How to set a pandas date range 

date_range = pd.date_range(start='2018-08-01', end='2019-07-17', freq='B') # freq='B' counts business days only, not weekends 
date_range
```

```python
# setting generated date range into the dataFrame

df_no_date.set_index(date_range, inplace = True)
df_no_date.head(10)
```

```python
# Plotting the graph...

%matplotlib inline 
df_no_date.Close.plot()
```

```python
# Specifying a certain range 
# date_range only can exclude weekends, not holidays 

df_no_date["2018-08-30":"2018-09-11"] # excludes the weekends 

```

```python
df_no_date["2018-09-01":"2018-09-11"].Close.mean()
```

```python
# daily prices
df_no_date.asfreq('D', method='pad') # adding the weekends as well by day, padding method will just carry on the values to the next 
```

```python
# weekly prices
df_no_date.asfreq('W', method='pad') 
```

```python
# monthly prices
df_no_date.asfreq('M', method='pad')
```
