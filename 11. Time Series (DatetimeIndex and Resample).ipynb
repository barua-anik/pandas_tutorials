# Resample

```python
# monthly avg of a close price 
df.Close.resample('M').mean()
```

```python
%matplotlib inline
df.Close.plot() # without resampling
```

```python
%matplotlib inline
df.Close.resample('M').mean().plot() # Month wise price 
```

```python
%matplotlib inline
df.Close.resample('M').mean().plot(kind="bar")
```
